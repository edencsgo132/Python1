```python
class NhanVien:
    def __init__(self, name, gender, hometown, departmentid, salary):
        self.name = name
        self.gender = gender
        self.hometown = hometown
        self.departmentid = departmentid
        self.salary = salary
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
# tạo cửa sổ ứng dụng

window = Tk()
window.title("Human resource management system ")
window.geometry('400x180')

#tạo list để lưu trữ toàn bộ các objects

data = []
# mô tả các function cho ứng dụng
# add/search/update/delete/reset

def add_nhan_vien():
    name_value = name.get()
    gender_value = gender.get()
    hometown_value = hometown.get()
    departmentid_value = departmentid.get()
    salary_value = salary.get()
    
    new_nv = NhanVien( name_value, gender_value,hometown_value, departmentid_value, salary_value)
    data.append(new_nv)
    
    print([item.name for item in data])
    
    # hiện thị thông báo đã add thành công hay chưa
    messagebox.showinfo('Thêm Nhân Viên', f"Thông tin của {name_value} được thêm vào thành công!")


    # xoá các giá trị ở các entry
    name.delete(0, END)
    gender.delete(0, END)
    hometown.delete(0, END)
    departmentid.delete(0, END)
    salary.delete(0, END)

# hàm xử lý tìm kiếm theo tên nhân viên

def search_nhan_vien():
    search_name_value = search_name.get()

    is_exist = False
    
    for nhanvien in data:
        if nhanvien.name == search_name_value:
            messagebox.showinfo("Tìm Kiếm Thông Tin", f'Tên: {nhanvien.name}, Giới tính: {nhanvien.gender}, Quê quán: {nhanvien.hometown}, Phòng ban: {nhanvien.departmentid}, Lương: {nhanvien.salary}')
            is_exist = True

    if is_exist == False:
        messagebox.showinfo("Tìm Kiếm Thông Tin", f'Nhân viên {search_name_value} không được tìm thấy!')
    
# hàm xử lý update thông tin nhân viên

def update_nhan_vien():
    name_value = name.get()
    gender_value = gender.get()
    hometown_value = hometown.get()
    departmentid_value = departmentid.get()
    salary_value = salary.get()
    
    
    is_exist = False
    
    for i in range (len(data)):
        if data[i].name == name_value:
            data[i].gender = gender_value
            data[i].hometown = hometown_value
            data[i].departmentid = departmentid_value
            data[i].salary = salary_value
            is_exist = True
            messagebox.showinfo("Cập nhật thông tin", f"Cập nhật thông tin của nhân viên {name_value} thành công!")
            break

    if is_exist == False:
        messagebox.showinfo("Cập nhật thông tin", 'Cập nhật thông tin không thành công!')

    # xoá các giá trị ở các entry
    name.delete(0, END)
    gender.delete(0, END)
    hometown.delete(0, END)
    departmentid.delete(0, END)
    salary.delete(0, END)



# delete theo tên nhan viên

def delete_nhan_vien():
    search_name_value = search_name.get()

    is_exist = False
    
    # xoá theo index
    for i in range (len(data)):
        if data[i].name == search_name_value:
            del data[i]
            messagebox.showinfo("Xoá Thông Tin", f'Thông tin tên nhân viên: {search_name_value} được xoá thành công!')
            is_exist = True
            break

    if is_exist == False:
        messagebox.showinfo("Xoá Thông Tin", 'Không tìm thấy thông tin!')

# hàm reset

def reset():
    name.delete(0, END)
    gender.delete(0, END)
    hometown.delete(0, END)
    departmentid.delete(0, END)
    salary.delete(0, END)
    search_name.delete(0, END)

# setup các widget
# tạo các label, có 5 label

Label(window, text = 'Tên nhân viên: ').grid(row =0, column =0)
Label(window, text = 'Giới tính: ').grid(row =1, column =0)
Label(window, text = 'Quê quán: ').grid(row =2, column =0)
Label(window, text = 'Phòng ban: ').grid(row =3, column =0)
Label(window, text = 'Lương nhân viên: ').grid(row =4, column =0)

# tạo các entry

name = Entry(window)
gender = Entry(window)
hometown = Entry(window)
departmentid = Entry(window)
salary = Entry(window)
search_name = Entry(window)


# tạo các button
add_btn = Button(window, text = 'Add',command= add_nhan_vien).grid(row =5, column =0)
delete_btn = Button(window, text = 'Delete', command= delete_nhan_vien).grid(row =5, column =1)
reset_btn = Button(window, text = 'Reset', command= reset).grid(row =5, column =2)
search_btn = Button(window, text = 'Search', command= search_nhan_vien).grid(row =6, column =1)
update_btn = Button(window, text = 'Update',command= update_nhan_vien).grid(row =6, column =2)


# setup vị trí các entry
name.grid(row =0, column =1)
gender.grid(row =1, column =1)
hometown.grid(row =2, column =1)
departmentid.grid(row =3, column =1)
salary.grid(row =4, column =1)
search_name.grid(row =6, column =0)


# Bước 6: chạy ứng dụng
window.mainloop()

```
